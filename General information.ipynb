````python
temp = input("What temperature is it?")
#type conver the string value entered into a number - an integer
temp = int(temp)

if temp >= 27:
    print("you should wear some shorts, short sleeves tshirts, slides and sunglusses")
elif temp >= 17:
    print("you shuld wear some jeans a long sleeve tshirt, trainers and a hat")
elif temp <= 10:
    print("you should wear a tracksuit, a jacket, gloves and trainers")
````
